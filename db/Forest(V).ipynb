```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("forestfires.csv")
```

```python
df
```

# 1 Pie

```python
sizes=[df['temp'].mean(),df['RH'].mean(),df['wind'].mean(),df['rain'].mean()]
```

```python
labels=['Temprature','Relative Humidity','Wind','Rain']
```

```python
colors=["red",'blue','yellow','green']
```

```python
explode=(0.1,0,0,0)
```

```python
plt.pie(sizes,explode=explode,colors=colors,labels=labels,autopct="%1.1f%%")
plt.show()
```

```python
y_bar=df.groupby('month')['temp'].mean()
y_bar
```

```python
x_bar=y_bar.index
x_bar
```

# 2 Bar

```python
plt.bar(x_bar,y_bar)

plt.xlabel("Month")
plt.ylabel("Temprature")
plt.title("Monthwise temp")
plt.show()
```

# 3 Hist

```python
plt.hist(df['temp'],bins='auto',color="orange",alpha=0.8)
plt.show()
```

# 4 Line plot

```python
x_line=df.iloc[:20,8]
y_line=df.iloc[:20,9]
```

```python
plt.plot(x_line,label='Temprature',marker="*",linestyle='dotted')
plt.plot(y_line,label="Relative humidity",marker="o",linestyle='dashed')
plt.legend()
plt.show()
```

# 5 Scatter

```python
plt.scatter(df['DMC'],df['DC'],color='red',marker='+')
# plt.ylim(0,500)
# The DMC is a numeric rating of the average moisture content of loosely compacted organic layers (duff layers) within the soil.
#The DC is a numeric rating of the average moisture content of deep organic layers in the soil.
```
